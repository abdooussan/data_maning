```python
#
import requests
from bs4 import BeautifulSoup
import csv
from itertools import zip_longest
#
import pandas as pd
```

# in this project we will scrap this [wep_page](https://wuzzuf.net/search/jobs/?q=python&a=hpb)

```python
# use requests to fetch the url
result = requests.get("https://wuzzuf.net/search/jobs/?q=python&a=hpb")
```

```python
#save the page content 
src = result.content
```

```python
#create the soup object
soup = BeautifulSoup(src, 'lxml')
```

Find all job titles





```python
job_titles_liste = []
job_titles = soup.find_all("h2",{"class":"css-m604qf"} )
for job_title in job_titles:
 job_titles_liste.append( job_title.find('a').text)

job_titles_liste
```

```python
links=[]

for job_title in job_titles:
 links.append(job_title.find("a").attrs['href'])
links
```

Find for every job the skills needs

```python
job_skills_liste = []
job_skills = soup.find_all("div",{"class":"css-y4udm8"})
for job_skill in job_skills:
 job_skills_liste.append( job_skill.text)
job_skills_liste
```

Company names

```python
company_names_liste = []
company_names = soup.find_all("a",{"class":"css-17s97q8"} )
for company_names in company_names:
 company_names_liste.append(company_names.text)
company_names_liste
```

```python
job_localisations_liste = []
job_localisations = soup.find_all("span",{"class":"css-5wys0k"} )
for job_localisation in job_localisations:
 job_localisations_liste.append( job_localisation.text)
job_localisations_liste
```

```python
print(len(job_titles_liste),len(job_localisations_liste),len(job_skills_liste),len(company_names_liste))
```

**Now we can create a data set, from this data**

```python
data = {"job_title":job_titles_liste,
        "company_name":company_names_liste,
        "job_localisation":job_localisations_liste,
        "job_skills":job_skills_liste,
        "job link": links
        }

df = pd.DataFrame(data)
print(df.shape)
df.head()
```


#create csv file

```python
file_list=[job_titles_liste,
           company_names_liste,
       job_localisations_liste,
       job_skills_liste,
       links]
exported = zip_longest(*file_list)
with open("/content/sample_data/scraping_data.csv", "w") as file:
    wr = csv.writer(file)
    wr.writerow(["job title","company name","localisation","skills","job link"])
    wr.writerows(exported)
```
